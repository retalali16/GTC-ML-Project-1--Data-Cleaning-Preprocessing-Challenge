# Handling The dataset

```python
df.drop(columns=['company'], inplace=True)
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df['agent'] = df['agent'].fillna(0)
```

```python
df['days_in_waiting_list'] = df['days_in_waiting_list'].fillna(0)
```

```python
df['required_car_parking_spaces'] = df['required_car_parking_spaces'].fillna(0)
```

```python
df['total_of_special_requests'] = df['total_of_special_requests'].fillna(0)
```

```python
df['customer_type'] = df['customer_type'].fillna(df['customer_type'].mode()[0])
```

```python
df['adr'] = df['adr'].fillna(df['adr'].median())
```

```python
df['reservation_status'] = df['reservation_status'].fillna(df['reservation_status'].mode()[0])
df['reservation_status_date'] = df['reservation_status_date'].fillna(df['reservation_status_date'].mode()[0])
```

```python
df.isnull().sum()
```

```python
above_1000 = df[df['adr'] > 1000]['adr']
print("Values in 'adr' above 1000:", above_1000.tolist())
```

```python
df['adr'] = df['adr'].clip(upper=1000)
```

```python
above_1000 = df[df['adr'] > 1000]['adr']
print("Values in 'adr' above 1000:", above_1000.tolist())
```

 #Convert arrival_date_month to ordered categorical

```python
import pandas as pd

months = ['January','February','March','April','May','June',
          'July','August','September','October','November','December']
df['arrival_date_month'] = pd.Categorical(df['arrival_date_month'], categories=months, ordered=True)
```

#Removed data leakage col

```python
df.drop(columns=['reservation_status', 'reservation_status_date'], inplace=True)
```

#One-Hot Encode categorical features

```python
categorical_cols = ['hotel', 'meal', 'market_segment', 'distribution_channel',
                    'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type']

categorical_cols = [col for col in categorical_cols if col in df.columns]

df_encoded = pd.get_dummies(df, columns=categorical_cols)
```

```python
df_encoded.columns.tolist()
```

#high cardinality in country column

```python
country_freq = df['country'].value_counts()
country_freq.tolist()
```

```python
country_proportions = country_freq / country_freq.sum()

country_proportions.tolist()
```

```python
df['country_freq'] = df['country'].map(country_proportions)
```

```python
df['country_freq']
```

# Feature Engneering

```python
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['is_family'] = ((df['children'] + df['babies']) > 0).astype(int)
```
